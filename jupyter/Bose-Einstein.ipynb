# Bose-Einstein vs. T

```python
%matplotlib notebook
# parola magica
import numpy as np
# le funzioni matematiche
import matplotlib.pyplot as P
# le funzioni grafiche
def n(E,mu):
    # questa Ã¨ la definizione di una routine (funzione), il testo che segue Ã¨ spostato a dx
    # il carattere : inizia la definizione della routine 
    # E e mu sono due valori in ingresso
    # il valore in uscita Ã¨ scritto dopo il comando return qui sotto
    # il cancelletto inizia il commento
    return 1./(np.exp(E-mu)-1)
# fine della funzione, ritorna a sn. D'ora in avanti n(2,1) calcola la funzione in questi due punti
# Se l'ingresso Ã¨ un vettore E numpy calcola come matlab l'intera funzione vettoriale 
# kT=1
fig,ax = P.subplots(figsize=(5,4))
# definisce una figura fig e un set di assi ax
mus = np.linspace(0.,-0.6,9)
# funzione di numpy, per vedere cosa fa digitare numpy linspace in google
for mu in mus:
    # anche i loop richiedono : e lo spostamento a dx
    E = np.linspace(0.01,1.2,1000)
    ax.plot(E,n(E,mu),label='{:.1f}'.format(mu))
    # genera il plot di una curva 
    # il loop produce tanti plot quanti i valori di mus e finisce con lo spostamento a sn
ax.set_xlabel('E')
# aggiunge l'etichetta dell'asse x, v. ad es.  pyplot x label
ax.set_ylabel(r'n$_B$(E,mu)')
# idem y
ax.legend(title=r'$\mu (k_BT=1)$',loc='right')
# aggiunge la legenda, v. ad es.  pyplot legend
P.show()
# chiude sempre questi file, e produce il grafico qui sotto
```
