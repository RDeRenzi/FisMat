# Potenziale Ven per la molecola HF (schizzo)
H ha configurazione $1s^1$ e $Z = 1$.

F ha configurazione $1s^2\,2s^2\, 2p^5$ quindi ha $Z = 9$, ma $N = 7$ elettroni nella shell esterna $n=2$ incompleta e $Z - N = 2$ elettroni nella shell completa $n'=1$. Di conseguenza ha $Z_{eff}\approx\frac {N-1} 2 = 3$.

```python
import matplotlib.pyplot as P
import numpy as np
```

```python
R = np.linspace(-1.5,1.5,101)
ZH,ZF = 1.,3.
nH,nF = 1,2
RH,RF = -0.5,0.5
VH = -(ZH/nH)**2/np.abs(R-RH)
VF = -(ZF/nF)**2/np.abs(R-RF)
Ven = VH+VF
fig,ax = P.subplots(figsize=(10,8))
ax.plot(R,VH)
ax.plot(R,VF)
ax.plot(R,Ven)
#ax.plot([-1,1],[0,0],'-k',lw=)
ax.set_xlabel(r'$\frac {r} {R}$')
ax.set_ylabel('E [Ry]')
ax.set_ylim(-10,0)
P.show()
```
