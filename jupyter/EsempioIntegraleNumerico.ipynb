## Esempio di integrale con quad

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as P
from scipy import stats
from scipy.integrate import quad

def integrando(x):
    return np.exp(-x)

 
def integrale(t):
    integral,err = quad(integrando,0.,t)
    return integral
vec_integrale = np.vectorize(integrale) 

fig, [ax1,ax2] = P.subplots(1,2,sharey='row')
x = np.linspace(0.,2.,101)
ax1.plot(x,vec_integrale(x),'r-')
ax2.plot(x,vec_integrale(4.*x),'b-')
ax1.plot([0.,2.],[0.,1.],'--')
ax1.set_xlim(0,1.99)
ax1.set_ylim(0,1)
ax2.set_xlim(0,2)
ax1.set_xlabel(r'$x$')
ax2.set_xlabel(r'$x$')
ax1.set_ylabel(r'$y$')
ax1.set_title('Esempio di integrazione (Google scipy quad)')
P.subplots_adjust(wspace=0.06)
P.show()
```
