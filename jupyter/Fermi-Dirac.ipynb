## Distribuzione di Fermi-Dirac

Nella simulazione si riproduce la distribuzione scalando energie e 
potenziale chimico all'energia termica

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as P
def n(E,mu):
    return 1./(np.exp(E-mu)+1)  # definizione della distribuzione di Fermi
# kB T = 1  (le energie sono riferite all'energia termica)
fig,ax = P.subplots(figsize=(5,4))
mus = np.array([0.01,0.3,1.,3.,10.,30.,100.]) # potenziale chimico da 1/100 a 100 volte kBT 
for mu in mus:
    E = np.linspace(0.01,1.5,150) # energia = E*mu, da 0 a 1.5 volte il potenziale chimico
    ax.plot(E,n(E*mu,mu),label='{:.1f}'.format(mu)) # 
    ax.plot(E,n(E*1000.,1000.),'k--',alpha=0.3,lw=0.5) # mu = 1000*kBT
ax.set_xlabel(r'$E/ \mu$')
ax.set_ylabel(r'n$_F(E,\mu)$')
ax.legend(title=r'$\mu/k_BT$')
P.show()
```

I casi di interesse sono quelli di elevanta densitÃ  di fermioni. Notare che l'energia di Fermi, $E_F =\mu(T=0)$, per i fermioni non interagenti di energia $\epsilon = \hbar^2 k^2/2m$, Ã¨ proporzionale alla loro densitÃ  alla potenza 2/3,

$$ E_F = \frac {\hbar^2} {2m} (3\pi^2 n)^{\frac 2 3} $$

e diventa molto grande per la densitÃ  di un elettrone per atomo (ad esempio per il rame  metallico $E_F/k_B = 60 000$, ossia nelle unitÃ  della simulazione mu = 300 a temperatura ambiente).

Il caso mu = 3, curva rossa, corrisponde per il rame alla temperatura di 20 000 K
