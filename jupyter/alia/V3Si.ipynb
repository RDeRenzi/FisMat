V$_3$Si presentation, why the offset?

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as P

p23 = np.genfromtxt('V3Si_23K.dat',delimiter=',') 
p18 = np.genfromtxt('V3Si_18K.dat',delimiter=',')
p = np.genfromtxt('v3si-PB.dat')#,delimiter=' ')

t23, a23, e23 = p23[:,0], p23[:,1], p23[:,2]
t18, a18, e18 = p18[:,0], p18[:,1], p18[:,2]
t, a = p[:,0]*1e6, p[:,1]


fig,ax = P.subplots(figsize=(8,6))

ax.errorbar(t23,a23,yerr=e23,fmt='ro',label='23 K')
ax.errorbar(t18,a18,yerr=e18,fmt='bo',label='18 K')
ax.plot(t,a*0.20,'r-',label='DFT cubic')

ax.set_xlabel(r'$t$  [$\mu$s]')
ax.set_ylabel('Asymmetry')

ax.legend()

P.show()
```
