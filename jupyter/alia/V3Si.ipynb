```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as P
from matplotlib import rcParams as rc
rc['font.serif'] = ['Carlito']#['Latin Modern Roman']
rc['font.family'] = "sans" #"serif"
rc['mathtext.fontset'] = 'cm'
# rc['text.usetex'] = True
rc['font.size'] = 16

# data extracted with mulab from # 3 - # 4, alpha = 0.995 (TF calibration)
# empiric fit gives 
a0 = 0.127+0.075+0.005 # errors are 0.005, 0.005 and 0.001 respectively

p23 = np.genfromtxt('V3Si_23K.dat',delimiter=',') 
p18 = np.genfromtxt('V3Si_18K.dat',delimiter=',')
p = np.genfromtxt('v3si-PB.dat')#,delimiter=' ')

t23, a23, e23 = p23[:,0], p23[:,1], p23[:,2]
t18, a18, e18 = p18[:,0], p18[:,1], p18[:,2]
t, a = p[:,0]*1e6, p[:,1]


fig,ax = P.subplots(figsize=(8,6))

ax.errorbar(t23,a23,yerr=e23,fmt='ro',label='23 K')
#ax.errorbar(t18,a18,yerr=e18,fmt='bo',label='18 K')
ax.plot(t,a*a0,'r-',label='DFT cubic')

ax.set_xlabel(r'$t$  [$\mu$s]')
ax.set_ylabel('Asymmetry')
ax.plot([0,10],[0,0],'k-',lw=0.5,alpha=0.5)
ax.legend()
ax.set_xlim(0,10)
ax.set_ylim(-0.01,0.22)


P.show()
```
