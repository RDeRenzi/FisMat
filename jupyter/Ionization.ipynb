## Ionizzazione

Modelli di schermo dell'elettrone nell'atomo

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as P

# kB T = 1  (le energie sono riferite all'energia termica)
fig,ax = P.subplots(figsize=(5,4))
Z = np.linspace(1,21,21)
n = np.array([1,1,2,2,2,2,2,2,2,2,3,3,3,3,3,3,3,3,4,4,4])
Nshell = np.array([0,0,2,2,2,2,2,2,2,2,10,10,10,10,10,10,10,10,18,18,18])
# numero di elettroni nella shell piena precedente
N = Z - Nshell 
# la shell piena scherma la carica nucleare per l'elettrone di valenza 
Nsubshell = np.array([0,0,2,2,4,4,4,4,4,4,10,10,12,12,12,12,12,12,18,18,20])
# include le subshell piene nello schermo
Nsub = Z - Nsubshell
E = np.array([13.59844,24.58741,5.39172,9.3227,8.29803,11.26030,14.53414,13.61806,17.42282,
              21.5646,5.13908,7.64624,5.98577,8.15169,10.48669,10.36001,12.96764,15.75962,
              4.34066,6.11316,6.5615])/13.59844
ax.plot(Z,E,'o')
ax.plot(Z,(Z/n)**2)
ax.plot(Z,(N/n)**2)
# conta solo le shell piene come schermo
ax.plot(Z,((N+1)/(2*n))**2)
# conta anche lo schermo parziale (campo medio) degli elettroni di valenza
ax.plot(Z,((Nsub+1)/(2*n))**2)
# conta schermo anche delle subshell piene e lo schermo parziale (campo medio) degli elettroni di valenza
ax.set_xlabel(r'$Z$')
ax.set_ylabel(r'$E_\mathrm{ion}$ [Ry]')
ymin,ymax = ax.get_ylim()
ax.set_ylim(0,5)#ymax)
P.show()
```

Il modello senza schermo produce energia di ionizzazione fuori scala che crescono con $Z^2$. Mitigato un po' dallo schelmo delle shell piene (corva verde).  Gli elettroni delle subshell p sono tanti e anche loro  schermano parzialmente la carica nucleare (curva rossa). Nei dati si vede un saltino in corrispondenza al riempimento della subshell s. Se quella subshell viene inclusa nel conto dello schermo si ottiene la curva violetta. Per le ulteriori deviazioni un campo medio brutale Ã¨ inadeguato.
